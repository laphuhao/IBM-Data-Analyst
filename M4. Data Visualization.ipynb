Import pandas module.

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY, conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
QUERY = """
SELECT ConvertedComp FROM master
"""
convertedcomp = pd.read_sql_query(QUERY, conn)
sns.displot(convertedcomp, kind="hist", bins=50);
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age FROM master
"""
age = pd.read_sql_query(QUERY, conn)
sns.boxplot(age);
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""
age_workweekhrs = pd.read_sql_query(QUERY, conn)
sns.scatterplot(data=age_workweekhrs, x='Age', y='WorkWeekHrs');
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
wca = pd.read_sql_query(QUERY, conn)

sns.scatterplot(data=wca, x=wca['WorkWeekHrs'], y=wca['CodeRevHrs'], size=wca['Age'], alpha=0.5);
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT * FROM sqlite_master
"""

pd.read_sql_query(QUERY, conn)
```

```python
QUERY = """
SELECT *, COUNT(*) AS Count FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY, conn)

data = df['Count']
labels = df['DatabaseDesireNextYear']

fig, ax = plt.subplots()
ax.pie(data, labels=labels, autopct='%1.1f%%', explode = (0.1, 0, 0, 0, 0), shadow = True);
```

In the list of most popular languages respondents wish to learn next year, what is the rank of Python?

```python
QUERY = """
SELECT *, COUNT(*) AS Count FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
"""
pd.read_sql_query(QUERY, conn)
```

How many respondents indicated that they currently work with SQL?

```python
QUERY = """
SELECT *, COUNT(*) AS Count FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY Count DESC
"""
pd.read_sql_query(QUERY, conn)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""
wca = pd.read_sql_query(QUERY, conn)

w_median = wca['WorkWeekHrs'].median()
c_median = wca['CodeRevHrs'].median()

plt.bar('Age group 30 to 35', w_median, color='r', label='WorkWeekHrs')
plt.bar('Age group 30 to 35', c_median, bottom=w_median, color='b', label='CodeRevHrs')
plt.legend()
plt.show();
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY, conn)

medians = df.groupby('Age')['ConvertedComp'].median()
medians.plot()
plt.ylabel('Median of Age Group from 45 to 60');
```

Based on the Line Chart of Age and median ConvertedComp for the age group 25-30, what kind of or correlation do you see between Age and median ConvertedComp?

```python
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 25 AND 30
"""
df = pd.read_sql_query(QUERY, conn)

medians = df.groupby('Age')['ConvertedComp'].median()
medians.plot()
plt.ylabel('Median of Age Group from 25 to 30');
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY, conn)
sns.barplot(data=df, x='Frequency', y='MainBranch', orient='h');
```
